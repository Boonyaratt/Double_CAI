```python
path = "E:\\Double-CAI\\Datasets\\Turi-Sampler-Data\\"
product = pd.read_csv(path + "products.csv")
promotions = pd.read_csv(path + "promotions.csv")
transactions = pd.read_csv(path + "transactions.csv")
promotion_transactions = pd.read_csv(path + "promotion_transactions.csv")
```

```python
merge_trans = pd.concat([transactions, promotion_transactions], axis=0, ignore_index=True)

merge_trans.tail()
```

```python
merge_trans_prom = pd.merge(merge_trans, promotions[['promo_id', 'product_scope']], how='left', left_on='promo_id', right_on='promo_id')
merge_trans_prom.tail()
```

```python
merge_trans_prom = pd.merge(
    merge_trans_prom,
    transactions[['transaction_id','branch_profile','province','zone']],
    on='transaction_id',
    how='left'
)
```
